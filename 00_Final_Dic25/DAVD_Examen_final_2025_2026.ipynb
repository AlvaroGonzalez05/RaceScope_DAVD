```python
# ============================================================
# 6. IDENTIFICACIÓN PROGRAMÁTICA DE OUTLIERS (IQR)
# ============================================================

def detect_outliers_iqr(series):
    """
    Entrada:
        series: variable numérica (pd.Series)
    Salida:
        máscara booleana indicando outliers según criterio IQR
    """
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return (series < lower) | (series > upper)


outlier_flags = {}

for col in cost_vars:
    outlier_flags[col] = detect_outliers_iqr(df[col])
    print(f"{col}: {outlier_flags[col].mean():.2%} de clientes identificados como outliers")

df_outliers = df.copy()

for col, mask in outlier_flags.items():
    df_outliers[f"outlier_{col}"] = mask
```

```python
# ============================================================
# 7. IMPACTO DE OUTLIERS EN EL COSTE MEDIO
# ============================================================

for col in cost_vars:
    mean_total = df[col].mean()
    mean_no_outliers = df.loc[~df_outliers[f"outlier_{col}"], col].mean()

    print(f"\nVariable: {col}")
    print("Coste medio total:", round(mean_total, 2))
    print("Coste medio sin outliers:", round(mean_no_outliers, 2))
    print("Reducción potencial (%):",
          round(100 * (mean_total - mean_no_outliers) / mean_total, 2))

```

```python
# ============================================================
# 3. LISTAS DE VARIABLES NUMÉRICAS Y CATEGÓRICAS
# ============================================================

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
numeric_exclude = ["person_id"]  # identificador
numeric_cols = [c for c in numeric_cols if c not in numeric_exclude]

categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()

print("\nVariables numéricas usadas en perfilado:")
print(numeric_cols)

print("\nVariables categóricas usadas en perfilado:")
print(categorical_cols)
```

```python
# ============================================================
# 4. PERFILADO NUMÉRICO DE OUTLIERS VS NO OUTLIERS
# ============================================================

numeric_profile = (
    df.groupby("high_cost_outlier")[numeric_cols]
    .mean()
    .transpose()
    .round(2)
)

print("\nPerfil numérico medio por grupo de coste (False = normal, True = outlier):")
print(numeric_profile)
```

```python
# ============================================================
# 5. PERFILADO CATEGÓRICO DE OUTLIERS VS NO OUTLIERS
# ============================================================

# Tabla de proporciones por categoría y grupo de outlier
categorical_profiles = {}

for col in categorical_cols:
    tab = (
        df.groupby("high_cost_outlier")[col]
        .value_counts(normalize=True)
        .rename("proportion")
        .reset_index()
    )
    categorical_profiles[col] = tab

# Ejemplo: mostrar algunas tablas de perfil categórico
print("\nPerfil categórico por grupo de coste para variables seleccionadas:")
for col in ["sex", "region", "urban_rural", "smoker", "alcohol_freq", "plan_type"]:
    if col in categorical_profiles:
        print(f"\nVariable categórica: {col}")
        print(categorical_profiles[col])
```

```python
# ============================================================
# 6. CLUSTERING DE CLIENTES OUTLIER (AGRUPACIÓN)
# ============================================================

outlier_mask = df["high_cost_outlier"]

cluster_numeric_vars = [
    "age",
    "bmi",
    "income",
    "risk_score",
    "claims_count",
    "chronic_count",
    "visits_last_year",
    "hospitalizations_last_3yrs",
    "days_hospitalized_last_3yrs",
    "ldl",
    "hba1c",
    "total_claims_paid",
]
cluster_numeric_vars = [c for c in cluster_numeric_vars if c in df.columns]

df_outliers = df.loc[outlier_mask, cluster_numeric_vars].dropna().copy()

print("\nNúmero de outliers usados en clustering:", df_outliers.shape[0])
print("Variables numéricas usadas en clustering:")
print(cluster_numeric_vars)

scaler = StandardScaler()
X_scaled_outliers = scaler.fit_transform(df_outliers)

# Selección de k fija (por simplicidad)
optimal_k = 3

kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df_outliers["cluster_outlier"] = kmeans.fit_predict(X_scaled_outliers)

print("\nDistribución de outliers por cluster:")
print(df_outliers["cluster_outlier"].value_counts())

cluster_outlier_profile = (
    df_outliers.groupby("cluster_outlier")[cluster_numeric_vars]
    .mean()
    .round(2)
)

print("\nPerfil medio numérico por cluster de outliers:")
print(cluster_outlier_profile)
```

```python
# ============================================================
# 7. ASIGNACIÓN DE CLUSTER DE OUTLIER AL DATASET COMPLETO
# ============================================================

df["cluster_outlier"] = np.nan
df.loc[df_outliers.index, "cluster_outlier"] = df_outliers["cluster_outlier"]

print("\nEjemplo de filas con cluster_outlier asignado:")
print(df.loc[df["high_cost_outlier"] == True, ["annual_medical_cost", "cluster_outlier"]].head())
```

```python
# ============================================================
# 8. REGRESIÓN LOGÍSTICA – PROBABILIDAD DE SER OUTLIER
# ============================================================

logistic_target = "high_cost_outlier"

logistic_numeric_features = [
    "age",
    "bmi",
    "income",
    "risk_score",
    "claims_count",
    "chronic_count",
    "visits_last_year",
    "hospitalizations_last_3yrs",
    "days_hospitalized_last_3yrs",
    "total_claims_paid",
]
logistic_numeric_features = [c for c in logistic_numeric_features if c in df.columns]

logistic_categorical_features = [
    "sex",
    "region",
    "urban_rural",
    "education",
    "marital_status",
    "employment_status",
    "smoker",
    "alcohol_freq",
    "plan_type",
    "network_tier",
]
logistic_categorical_features = [c for c in logistic_categorical_features if c in df.columns]

logistic_features = logistic_numeric_features + logistic_categorical_features

df_logistic = df.dropna(subset=[logistic_target] + logistic_features).copy()

X_log = df_logistic[logistic_features]
y_log = df_logistic[logistic_target].astype(int)

preprocessor_log = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), logistic_numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), logistic_categorical_features),
    ]
)

logistic_model = LogisticRegression(max_iter=1000)

logistic_clf = Pipeline(steps=[
    ("preprocess", preprocessor_log),
    ("model", logistic_model),
])

X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(
    X_log, y_log, test_size=0.3, random_state=42, stratify=y_log
)

logistic_clf.fit(X_train_log, y_train_log)

y_pred_log = logistic_clf.predict(X_test_log)

log_acc = accuracy_score(y_test_log, y_pred_log)
log_prec = precision_score(y_test_log, y_pred_log)
log_rec = recall_score(y_test_log, y_pred_log)
log_f1 = f1_score(y_test_log, y_pred_log)

print("\nMétricas del modelo de regresión logística (alto coste vs resto):")
print("Accuracy :", round(log_acc, 3))
print("Precisión:", round(log_prec, 3))
print("Recall   :", round(log_rec, 3))
print("F1       :", round(log_f1, 3))

# Importancia de variables (coeficientes del modelo lineal en el espacio preprocesado)
log_reg = logistic_clf.named_steps["model"]
log_pre = logistic_clf.named_steps["preprocess"]

ohe = log_pre.named_transformers_["cat"]
cat_feature_names = ohe.get_feature_names_out(logistic_categorical_features)

feature_names_log = np.concatenate([logistic_numeric_features, cat_feature_names])

coef_log = log_reg.coef_.ravel()

importance_log_df = (
    pd.DataFrame({"variable": feature_names_log, "coeficiente": coef_log})
    .sort_values("coeficiente", ascending=False)
)

print("\nImportancia (coeficientes) de las variables en la regresión logística:")
print(importance_log_df)
```
